```python
#Question 1
#import the necessary library
import matplotlib.pyplot as pl
#open and read the text file and input every element into a list t
f = open('Map4.txt','r')
t = f.readlines()
f.close()
#determine the size of the map
M = int(t[0].rstrip())
N = int(t[1].rstrip())
#create the map as a list of lists (matrix)
RM = range(0,M)
RN = range(0,N)
Map = []
c = 2
for i in RM:
    row = []
    for j in RN:
        row.append(int(t[c]))
        c += 1
    Map.append(row)
#create variables for the current and previous position and set them to 0
cr = 0
cc = 0
pr = 0
pc = 0
#create a list of possible movements
move = [-1, 0, 1]
#set a boolean variable to continue the journey to True
cont = True
#create a list of row and column variables for the path taken
row = [0]
column = [0]
#continue moving while the boolean variable is true
while cont:
    #set the variable to False
    cont = False
    #check all rows and columns around the current one
    for i in move:
        for j in move:
            r = cr + i
            c = cc + j
            #if the analysed position has number 0 and it is not the previous one, then the journey must continue and the path saved into the record lists
            if Map[r][c] == 0 and not (r == cr and c == cc) and not (r == pr and c == pc):
                pr = cr
                pc = cc
                cr = r
                cc = c
                cont = True
                row.append(r)
                column.append(c)
#plot the path and the final position
pl.plot(row, column, linestyle = '--', c = 'Black')
pl.scatter(cr, cc, marker = 'x', c = 'Red', s = 200)
pl.xlim(0)
pl.ylim(0)
#print the map
print('Treasure found in ' + str(cr) + ', ' + str(cc) + '!')
```
