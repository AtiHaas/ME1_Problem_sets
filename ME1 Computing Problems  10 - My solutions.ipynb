```python
# TASK A

def factorial(n):
    if n == 1:
        return n
    else:
        return n*factorial(n-1)
    
    
import matplotlib.pyplot as pl
import numpy as np

N = 2*int(input("How many terms do you want?"))

x = []
y = []

for i in np.arange(0, 12.56, 0.01):
    x.append(i)

for x_ in x:
    y1 = 0
    for i in range(1,N+1,2):
        y1 += (-1)**(i//2)*(x_**i)/factorial(i)
    y.append(y1)
                    
                    
pl.plot(x,y)





        
```
